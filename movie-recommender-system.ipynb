```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'

this is a string and that is why it is not working. So we first have to convert this string of list into list.
For this we have a module called ast in python which has a function literal_eval() which does this work.

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# therefore I am creating the above function again which implements literal_eval() function
def convert(obj):
    L = []               # creating a list
    for i in ast.literal_eval(obj):
        L.append(i['name'])         # appending the value to the key 'name' from the dictionary to the list
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

the above function is working fine

###### here we will take top 3 actors and not all the cast

```python
# {"cast_id": 242, "character": "Jake Sully", "credit_id": "5602a8a7c3a3685532001c9a", "gender": 2, "id": 65731, "name": "Sam Worthington", "order": 0}
# in this "Jake Sully" is the name of character in the movie and "Sam Worthington" is the actual name of the actor
# therefore, we will take the the actual name. The same above function is repeated with some minor changes

def convert3(obj):
    L = []               # creating a list
    counter = 0            # to take maximum of 3 casts
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])         # appending the value to the key 'name' from the dictionary to the list
            counter+=1
        else:
            break
    return L
```

# Vectorization
Here we will check the similarities between tags so that we could recommend similar movies.

For this, we will combine all the tags and create a large_text. For example, large_text = tag1 + tag2 + ..... +tag5000

Now, from this large_text, we will take out lets say 5000 most common words and create vectors with them.

We will use CountVectorizer from Sklearn.feature_extraction.text for this. Here we will use the stopwords functionality. Stopwords are the words which are used in the sentence formation. For example: are, is, in, the, etc.
We will neglect these stopwords

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return" ".join(y)
```

```python
stem('loved loving love')
```

```python
stem('dance dancing danced')
```

```python
new_df['tags'].apply(stem)  # performing stemming
```

```python
# storing the above result in the same column
new_df['tags'] = new_df['tags'].apply(stem)
```

now to remove words which have similar meaning
for example ['loved','loving','love']    =>  ['love','love','love']
we will use stemming for this

we will use nltk library for this.


we have created a stem function above for this thing.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')    # max_features=5000 these are the words taken out from large_text
```

```python
cv.fit_transform(new_df['tags']).toarray()      # explicitly converting tags to numpy array as cv returns a sparse matrix
```

```python
# saving the above result in a new variable
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
# to check 5000 most frequent words
cv.get_feature_names_out()
```

##### now we have 4806 vectors which have 5000 numbers each
##### we need to calculate the distance between each and every movie to each and every other movie 
##### the more the distance, lesser the similarity between those 2 movies
##### here we will not calculate the euclean distance, instead we will calculate the cosine distance
##### i.e., euclean distance is sqrt((x2-x1)^2 - (y2-y1)^2) and cosine distance is the angle between the two vectors
### Note: If you are working with high dimensional data, as in now as we have 4806x5000, euclean distance fails. It is not a reliable measure for higher dimensional data.

##### for cosine similarity, we have a function in sklearn.metrics.pairwise known as cosine_similarity which ranges between 0 to 1. If it is 1 then similarity is more else not.

```python
from sklearn.metrics.pairwise import cosine_similarity    
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
# storing all these distances in a variable
similarity = cosine_similarity(vectors)
```

now the next step is to create a function which takes movie as an input and returns 5 similar movies.

To do this, first a movie is taken as input. Then we have to find the index of the movie in the data. Corresponding to that index, we will go to the similarity matrix which contains the distance of that movie with other movies. Then we will sort the similarity score in descending order i.e., more similar movies will come in the beginning. (more the similarity score, more similar the movie). Then we will fetch the top 5 movies and display them.

```python
# we have to hold the positions of distance as after sorting, the indexes will change and we won't be able to find out which score at which index belongs to which movie
# for this we will use enumerate() function which returns an object which when converted to list given the following result
list(enumerate(similarity[2]))
```

```python
# now we can sort the above data
sorted(list(enumerate(similarity[2])),reverse=True,key=lambda x:x[1])    # if key was not written, sorting would have been done on the basis of index values
```

```python
# fetching top 5 movies
sorted(list(enumerate(similarity[2])),reverse=True,key=lambda x:x[1])[1:6]
```

## Now to implement all these functionalities discussed above, we will create a function

```python
# function which takes movie as an input and returns 5 similar movies
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)     # new_df.iloc[i[0]].title will fetch 5 movies name from the data. i[0] will give the top 5 movies index and corresponding to that index we will fetch the movie name
```

```python
recommend('Avatar')
```

```python
recommend('The Avengers')
```

```python
recommend('Batman')
```
